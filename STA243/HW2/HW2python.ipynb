```python
# Implementing Gradient Descent 
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
train_scale = scale.fit_transform(train_sub)
display(train_scale)
```

```python
# a
train_noInt = train[["bedrooms", "bathrooms", "sqft_living", "sqft_lot"]]

def graddescOLS(X, y, eta = None, tau = 1e-6):
    # Standardize the data
    X = StandardScaler().fit_transform(X)

    # Add column of 1s for intercept
    X = sm.add_constant(X)

    # Eta is stepsize, Tau is error
    if eta is None: 
        # eta = 2 / \kappa + 1 
        eta = 2 / (np.linalg.cond(X.T @ X) + 1)

    # Initialization of beta and a gradient vector
    n,p = X.shape[0], X.shape[1]
    beta = np.zeros(p)
    grad = np.ones(p) # initilization

    while np.linalg.norm(grad) > tau:
    # Formula for gradient
        grad =2/n * X.T @ (X@beta - y)
        beta = beta - eta * grad #X.T @ (X@beta - y)

    return beta

beta_gd_noInt = graddescOLS(train_noInt, train_price)
yhat_betaGD = sm.add_constant(scale.fit_transform(train_noInt)) @ beta_gd_noInt
r2_score(y_pred=yhat_betaGD, y_true=train_price)
```
