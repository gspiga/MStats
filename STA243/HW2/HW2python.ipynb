```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from IPython.display import display

import statsmodels.api as sm
```

```python
train = pd.read_csv("train.data.csv").drop("Unnamed: 0", axis = 1)
display(train)
train_sub = train[["bedrooms", "bathrooms", "sqft_living", "sqft_lot"]]
display(train_sub)
train_price = train["price"]

# Build regression model
reg = LinearRegression().fit(train_sub, train_price)
print(r2_score(train_price, reg.predict(train_sub)))

#r2_score(train_sub, train_price)
#train_sub = sm.add_constant(train_sub)
#result = sm.OLS(train_price, train_sub).fit()
#print(result.rsquared)

test = pd.read_csv("test.data.csv")
test_sub = test[["bedrooms", "bathrooms", "sqft_living", "sqft_lot"]]
test_price = test["price"]

print(r2_score(test_price, reg.predict(test_sub)))
```
