## Problem 3
Prove that if the function $f: \mathbb{R}^n \rightarrow \mathbb{R}$ is differentialble, the gradient of $f$ at a point is either zero, or perpendicular to the level set of $f$ at that point. 

We define the gradient of $f: \mathbb{R}^n \rightarrow \mathbb{R}$ as the following vector:
$$
\begin{align*}
\nabla f = \left(\frac{\partial f}{\partial x_1}, \frac{\partial f}{\partial x_2}, \ldots, \frac{\partial f}{\partial x_n}\right)
\end{align*}
$$
and level set $L(c) = \{x \in \mathbb{R}^n ; f(x) = c\}$ where $c = f(\gamma)$ and $\gamma$ is a point in $\mathbb{R}^n$. 

For level set $L(c)$, we can define a vector $t$ which is tangent at point $\gamma$. The directional derivative of $f$ at this point $\gamma$ can be defined by the following dot product. 

$$
\begin{align*}
D_{\gamma}f &= \frac{\nabla f(\gamma) \ t}{||t||}
\end{align*}
$$

To show that two vectors are perpendicular, we must show their dot product equals zero. 

$$
\begin{align*}
0 &= \frac{\nabla f(\gamma) \ t}{||t||}
\end{align*}
$$

The above can only be true when $t$ is the trivial $\vec{0}_n$, $\nabla f(\gamma) = 0$, or the vector $t$ is perpendicular to $\nabla f(\gamma)$ at point $\gamma$.

Finding the stationary points:
$$
\begin{align*}
f(x) &= x^2 - x^4 \\
f'(x) &= 2x - 4x^3 \\
0 &= 2x(1 - 2x^2) \\
0 = 2x \ &\text{and} \ 0 = 1 - 2x^2 \\ 
0 = x \ &\text{and} \ x = \pm \sqrt{\frac{1}{2}} \\ 
\end{align*}
$$

We can find the second derivative for minima: 
$$
\begin{align*}
f''(x) &= 2 - 12x^2 \\
f''(0) &= 2 \\
f''(\sqrt{\frac{1}{2}}), f''(-\sqrt{\frac{1}{2}}) &= -4 \\
\end{align*}
$$

The point $x = 0$ is a local minimum while the points $x = \sqrt{\frac{1}{2}}, -\sqrt{\frac{1}{2}}$ are local maximums.  From the graph above, we can see that because the function approaches $-\infty$ and $\infty$ in both directions of $x$, there is no global minimum.
```python
def function(n):
    i = 1
    while (i <= n):
        j = 1
        while (j <=n):
            g(i,j)

            k =1
            while (k <= n):
                h(i, j, k)
                k = k*3
            j = j+i
        i = i +1
```

For the function $g$, we have the first while loop goes from $i = 1 \ldots n$. However, we must consider the second loop more carefully, since the updating step of $j$ is dependent on $i$. 

- When $i = 1$, second loop is executed $n$ times. when $i = 2$, the second loop is only executed $\frac{n}{2}$ times, $i=3$ corresponds to $\frac{n}{3}$ times and so on. This pattern is identical to that of a harmonic series, which asymptotically converges to $\log(n)$ Thus the first two loops are $O(n \log(n))$ for the function $g$. 
- For the function $h$, $k$ is always being multiplied by 3 until it is less than or equal to $n$. Since $k=1$ at initialization, we can solve the following for the $l$ amount of loops:
$$
\begin{align*}
n &= 3^l \\ 
\log(n) &= l \log(3) \\ 
\frac{\log(n)}{\log(3)} &= l \\
\end{align*}
$$

Since $\log(3)$, we have the function $h$ has time complexity $O(\log(n))$, leaving for a total time complexity for the entire function as $O(n \log^2(n))$.
